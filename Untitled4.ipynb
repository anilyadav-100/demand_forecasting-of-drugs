```python
# ADF Test to check stationarity
def adf_test(series):
    result = adfuller(series.dropna())
    print(f"ADF Statistic: {result[0]}")
    print(f"p-value: {result[1]}")
    if result[1] < 0.05:
        print("Data is stationary ")
    else:
        print("Data is NOT stationary ")

# Test stationarity
adf_test(new_df['Demand'])

# If data is not stationary, apply first differencing
df['Demand_diff'] = df['Demand'].diff().dropna()
adf_test(df['Demand_diff'])
```

```python
from transformers import TFAutoModel, AutoTokenizer
import numpy as np
```

```python
tokenizer = AutoTokenizer.from_pretrained('bert-base-uncased')
bert_model = TFAutoModel.from_pretrained('bert-base-uncased')
```

```python
def get_bert_embedding(row):
    """Generate a single BERT embedding for combined text features using a separator token."""
    # Combine the text fields using the `[SEP]` token
    combined_text = f"{row['Event']} [SEP] {row['Market_News']} [SEP] {row['Press_Release']}"
    
    if pd.isna(combined_text):  # Handle missing values
        return np.zeros(768)  # 768-dimensional zero vector
    
    # Tokenize and encode input with padding/truncation
    inputs = tokenizer(combined_text, return_tensors='tf', padding=True, truncation=True, max_length=100)
    
    # Pass through BERT model
    outputs = bert_model(**inputs)
    
    # Extract the [CLS] token's embedding (single 768-d vector)
    return outputs.last_hidden_state[:, 0, :].numpy().flatten()

# Apply BERT encoding on all rows
new_df['BERT_Embedding'] = new_df.apply(get_bert_embedding, axis=1)

```

```python

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler









y = new_df['Demand'].values  # Shape: (num_samples,)

# Normalize input features
scaler_X = MinMaxScaler()
X_scaled = scaler_X.fit_transform(X)

scaler_y = MinMaxScaler()
y_scaled = scaler_y.fit_transform(y.reshape(-1, 1))

# Reshape X for LSTM (samples, time_steps, features)
X_lstm = X_scaled.reshape(X_scaled.shape[0], 1, X_scaled.shape[1])  # Time step = 1 for single-step prediction

# Split into train and test sets (80-20 split)
train_size = int(0.8 * X_lstm.shape[0])
X_train, X_test = X_lstm[:train_size], X_lstm[train_size:]
y_train, y_test = y_scaled[:train_size], y_scaled[train_size:]

# LSTM Model
model = Sequential([
    LSTM(256, return_sequences=True, input_shape=(1, X_lstm.shape[2])),
    Dropout(0.2),
    LSTM(128, return_sequences=False),
    Dropout(0.2),
    Dense(64, activation='relu'),
    Dense(32, activation='relu'),
    Dense(1)  # Single output for demand prediction
])

# Compile model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Train the model
model.fit(X_train, y_train, epochs=50, batch_size=32, validation_data=(X_test, y_test), verbose=1)

# Make predictions
y_pred = model.predict(X_test)

# Convert predictions back to original scale
y_pred_original = scaler_y.inverse_transform(y_pred)

# Evaluate model
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f"Test MAE: {test_mae}")

# Example: Forecast for the next day
next_day_input = X_lstm[-1].reshape(1, 1, X_lstm.shape[2])  # Take the last available input
next_day_prediction = model.predict(next_day_input)
next_day_demand = scaler_y.inverse_transform(next_day_prediction)
print(f"Forecasted Demand for Next Day: {next_day_demand[0][0]}")
```
